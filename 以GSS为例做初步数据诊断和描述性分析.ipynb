```python
#引入第三方库并进行基础设置
import pandas as pd
import matplotlib.pyplot as plt

#取消科学计数法
pd.set_option("display.float_format", lambda x:"%.1f"% x)

sample_data=pd.read_csv("F:\GSS Project\GSS Project\2000_2012.csv") 
```

### Missing values count   # 统计丢失数据

```python
#对GSS表进行缺失值诊断
pd.options.display.max_rows = 300
sample_data_na_description_dict={}
for col in (sample_data.columns.values.tolist()):
    cnt_na=0
    cnt=0
    for value in sample_data[col]:
        if pd.isna(value)==True:
            cnt_na+=1
        else:
            cnt+=1
    sample_data_na_description_dict[col]={"Non missing value":cnt,"missing value":cnt_na,"sample size":cnt+cnt_na,"missing value ratio":cnt_na/(cnt+cnt_na)}

sample_data_na_description_table=pd.DataFrame(sample_data_na_description_dict).T
sample_data_na_description_table["missing value ratio"]=sample_data_na_description_table["missing value ratio"].apply(lambda x: '%.2f%%' % (x*100))

sample_data_na_description_table
```

## Proportion of statistical category variable 统计分类变量

```python
pd.options.display.max_rows = 500
discrete_freature_list=[]
continuous_freature_list=[]
for col in (sample_data.columns.values.tolist()[1:]):
    if sample_data[col].dtypes=="object":
        discrete_freature_list.append(col)
    else:
        continuous_freature_list.append(col)
#定义类别变量描述函数
def get_discrete_freature_description(discreet_feature):
    describe_dict={}
    lst=list(sample_data[discreet_feature].unique())
    for item in lst:
        cnt=0
        for value in sample_data[discreet_feature]:
            if item==value:
                     cnt+=1
            elif pd.isna(item)==True and pd.isna(value)==True:
                     cnt+=1
            else:
                     pass
        describe_dict[item]=cnt
    describe_table=pd.DataFrame.from_dict(describe_dict,orient='index')
    describe_table.reset_index(inplace=True)
    describe_table.rename(columns={0:"classified sample size","index":"classification value"}, inplace = True)
    describe_table["Variable name"]=discreet_feature
    describe_table["total sample"]=sum(describe_dict.values())
    describe_table["proportion"]=describe_table["classified sample size"]/describe_table["total sample"]
    describe_table["proportion"]=describe_table["proportion"].apply(lambda x: '%.2f%%' % (x*100))
    describe_table=describe_table[["Variable name","classification value","total sample","classified sample size","proportion"]] 
    return describe_table

discrete_freature_description_table=pd.DataFrame(columns=["Variable name","classification value","total sample","classified sample size","proportion"])
for feature in discrete_freature_list:
    describe_table=get_discrete_freature_description(feature)
    discrete_freature_description_table=pd.concat([discrete_freature_description_table,pd.DataFrame(describe_table)],axis=0)

discrete_freature_description_table.set_index(["Variable name","classification value"],inplace=True)

discrete_freature_description_table
```
